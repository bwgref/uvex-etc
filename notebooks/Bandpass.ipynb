```python
import uvex_etc
import numpy as np
import astropy.units as u
import matplotlib.pyplot as plt
from uvex_etc import filters

from importlib import reload
reload(filters)
```

```python
nuv_band = filters.nuv_bandpass()
fuv_band = filters.fuv_bandpass()
lss_wave = filters.lss_wave()
lss_band = filters.lss_bandpass()
```

```python
wave = np.arange(1000, 3000) * u.AA
```

```python
ax = plt.figure().subplots()
nuv_resp = nuv_band.__call__(wave)
fuv_resp = fuv_band.__call__(wave)

ax.plot(wave, nuv_resp, label= 'NUV')

ax.plot(wave, fuv_resp, label= 'FUV')

lss_resp = lss_band.__call__(lss_wave)

ax.plot(lss_wave, lss_resp, label= 'LSS')
ax.set_xlabel(wave.unit)
ax.set_xlim(1000, 3000)
ax.set_ylabel('Throughput')
ax.legend()
plt.show()


```
